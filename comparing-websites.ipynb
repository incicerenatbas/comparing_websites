****Analysis Fandango's site:

```python
fandango = pd.read_csv("/kaggle/input/fandango-scrapecsv/fandango_scrape.csv")
```

```python
fandango.head()
```

```python
fandango.info()
```

```python
fandango.describe()
```

```python
#Explore the relationship between movie votes and rating:
plt.rcParams["figure.figsize"]=(14,8)
plt.scatter(fandango["RATING"],fandango["VOTES"],
            color="pink")
plt.xlabel("RATING")
plt.ylabel("VOTES")
plt.show()
```

```python
#See the relationship between all columns:
fandango.corr()
```

```python
#Add 'Year' column:
fandango["YEAR"]=fandango["FILM"].astype(str).str[-5:-1]
fandango
```

```python
#Find movies per year:
fandango.groupby("YEAR")["FILM"].count()
```

```python
#Visualize the number of movies per year with a plot:
sns.countplot(data=fandango["FILM"],x=fandango["YEAR"])
plt.xlabel="YEAR"
plt.ylabel="count"
plt.show()
```

```python
# The 10 movies with the most votes:
fandango.nlargest(10,"VOTES")
```

```python
#How many movies got 0 votes:
fandango[fandango["VOTES"] == 0].shape[0]
```

```python
#Create dataframe only of reviewed movies by removing movies with zero votes: 
fandango=fandango[fandango["VOTES"]!=0]
fandango
```

```python
#Keep the differences between STARS and RATING in this column by adding a new column named 'STARS_DIFF':
fandango["STARS_DIFF"]= fandango["STARS"]-fandango["RATING"]
fandango
```

```python
#Count and plot how many times the data in the STARS_DIFF column has occurred:
fandango["STARS_DIFF"].count()
sns.countplot(data=fandango, x="STARS_DIFF")
plt.xticks(rotation=45)
plt.show()
```

```python
#As we can see from the plot above, there is even a movie with a STARS_DIFF of 1. Bring this very heavily corrupted film and let's examine it:
fandango[fandango["STARS_DIFF"]==1]
```

****Let's Compare Fandango Ratings with Other Sites:

```python
all_sites = pd.read_csv("/kaggle/input/all-sites-scorescsv/all_sites_scores.csv")
```

```python
all_sites.head()
```

```python
all_sites.info()
```

```python
all_sites.describe()
```

```python
#Review Rotten Tomatoes first.
#There is a column called RottenTomatoes with ratings from real reviewers and a column called RottenTomatoes_User with ratings from users. 
#Let's plot these columns.
sns.scatterplot(x=all_sites["RottenTomatoes"],y=all_sites["RottenTomatoes_User"])
```

```python
#Create a column named Rotten DIFF and this column contains RottenTomatoes-RottenTomatoes_User information:
all_sites["Rotten_DIFF"]=all_sites["RottenTomatoes"]-all_sites["RottenTomatoes_User"]
all_sites.head()
```

```python
#RT Critics score minus RT User Score:
sns.histplot(all_sites["Rotten_DIFF"], bins=25, kde=True)
plt.title("RT Critics score minus RT User Score")
plt.legend()
plt.show()
```

```python
#According to average, bring in 5 movies that users liked more than critics:
all_sites.nsmallest(5,"Rotten_DIFF")
```

```python
#Now just the opposite, bring 5 movies that users dislike even more than the critics:
all_sites.nlargest(5,"Rotten_DIFF")
```
